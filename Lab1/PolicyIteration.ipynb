```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Oct  9 22:03:20 2019

@author: tiantong

9261299999286920541198
"""
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import time
'''
Policy Iteration
'''
X = 8
Y = 8
H = 12 # number of heading directions

FORWARD = 1
BACKWARD = -1
STILL = 0
LEFT_TURN = -1
RIGHT_TURN = 1
NO_TURN = 0

def initialize(X, Y):
    '''
    initialize the states set and actions set.
    inputs: X: the width of the grid; Y: the height of the grid.
    outputs: S: states set, A: actions set.
        
    '''
    S = [] # states set
    for x in range(X):
        for y in range(Y):
            for h in range(H):
                S.append((x, y, h))
    A = [] # actions set  
    for move in [STILL, FORWARD, BACKWARD]:
        if move != STILL:
            for turn in [LEFT_TURN, RIGHT_TURN, NO_TURN]:
                A.append((move, turn))
        else:
            A.append((STILL, NO_TURN))
    return S, A

def transitionProb(s, a, s_prime, pe=0.0):
    '''
    returns the probability psa(s′) given inputs pe, s, a, s′
    inputs: s: current state (x, y, h)
            a: action (move, turn)
            s_prime: next state (x', y', h')
            pe: error probobility
    outputs: p: the probability psa(s′) given inputs pe, s, a, s′
    '''
    if pe < 0 or pe > 0.5:
        raise ValueError('The error probability must be between 0 and 0.5')
    possible_move_dirs = {}
    # [0, 1]: moving towards +x direction; [0, -1]: moving towards -x direction
    # [1, 0]: moving towards +y direction; [-1, 0]: moving towards -y direction
    # possible_move_dir[*] contains three entries:
    # (moving direction--if moving forward, heading without pre-rotation, probability without pre-rotation)
    # (moving direction--if moving forward, heading with pre-rotation1, probability with pre-rotation1)
    # (moving direction--if moving forward, heading with pre-rotation2, probability with pre-rotation2)
    possible_move_dirs[0] = [([0, 1], 0, 1-2 * pe), ([0, 1], 1, pe), ([0, 1], 11, pe)]
    possible_move_dirs[1] = [([0, 1], 1, 1-2 * pe), ([1, 0], 2, pe), ([0, 1], 0, pe)]
    possible_move_dirs[2] = [([1, 0], 2, 1-2 * pe), ([1, 0], 3, pe), ([0, 1], 1, pe)]
    possible_move_dirs[3] = [([1, 0], 3, 1-2 * pe), ([1, 0], 4, pe), ([1, 0], 2, pe)]
    possible_move_dirs[4] = [([1, 0], 4, 1-2 * pe), ([0, -1], 5, pe), ([1, 0], 3, pe)]
    possible_move_dirs[5] = [([0, -1], 5, 1-2 * pe), ([0, -1], 6, pe), ([1, 0], 4, pe)]
    possible_move_dirs[6] = [([0, -1], 6, 1-2 * pe), ([0, -1], 7, pe), ([0, -1], 5, pe)]
    possible_move_dirs[7] = [([0, -1], 7, 1-2 * pe), ([-1, 0], 8, pe), ([0, -1], 6, pe)]
    possible_move_dirs[8] = [([-1, 0], 8, 1-2 * pe), ([-1, 0], 9, pe), ([0, -1], 7, pe)]
    possible_move_dirs[9] = [([-1, 0], 9, 1-2 * pe), ([-1, 0], 10, pe), ([-1, 0], 8, pe)]
    possible_move_dirs[10] = [([-1, 0], 10, 1-2 * pe), ([0, 1], 11, pe), ([-1, 0], 9, pe)]
    possible_move_dirs[11] = [([0, 1], 11, 1-2 * pe), ([0, 1], 0, pe), ([-1, 0], 10, pe)]
    
    P = {} # all the probabilities
    if a[0] == FORWARD:
        for possible_move_dir in possible_move_dirs[s[2]]:
            x_hypo = s[0] + possible_move_dir[0][0]
            x = x_hypo if (x_hypo <= X - 1 and x_hypo >= 0) else s[0] # out of the grid cell
            y_hypo = s[1] + possible_move_dir[0][1]
            y = y_hypo if (y_hypo <= Y - 1 and y_hypo >= 0) else s[1] # out of the grid cell
            h = (possible_move_dir[1] + a[1]) % 12
            P[(x, y, h)] = possible_move_dir[2]
    
    elif a[0] == BACKWARD:
        for possible_move_dir in possible_move_dirs[s[2]]:
            x_hypo = s[0] - possible_move_dir[0][0]
            x = x_hypo if (x_hypo <= X - 1 and x_hypo >= 0) else s[0] # out of the grid cell
            y_hypo = s[1] - possible_move_dir[0][1]
            y = y_hypo if (y_hypo <= Y - 1 and y_hypo >= 0) else s[1] # out of the grid cell
            h = (possible_move_dir[1] + a[1]) % 12
            P[(x, y, h)] = possible_move_dir[2]
    else:
        P[s] = 1 # stay still with prob. = 1
    if s_prime in P.keys():
        return P[s_prime]
    else:
        return 0.0
    
def nextState(s, a, pe):
    '''
    returns the next state given current state, action, and error probability
    inputs: s: current state (x, y, h)
            a: action (move, turn)
            pe: error prob.
    outputs: All possible next state with probability
    '''
    if pe < 0 or pe > 0.5:
        raise ValueError('The error probability must be between 0 and 0.5')
    possible_move_dirs = {}
    # [0, 1]: moving towards +x direction; [0, -1]: moving towards -x direction
    # [1, 0]: moving towards +y direction; [-1, 0]: moving towards -y direction
    # possible_move_dir[*] contains three entries:
    # (moving direction--if moving forward, heading without pre-rotation, probability without pre-rotation)
    # (moving direction--if moving forward, heading with pre-rotation1, probability with pre-rotation1)
    # (moving direction--if moving forward, heading with pre-rotation2, probability with pre-rotation2)
    possible_move_dirs[0] = [([0, 1], 0, 1-2 * pe), ([0, 1], 1, pe), ([0, 1], 11, pe)]
    possible_move_dirs[1] = [([0, 1], 1, 1-2 * pe), ([1, 0], 2, pe), ([0, 1], 0, pe)]
    possible_move_dirs[2] = [([1, 0], 2, 1-2 * pe), ([1, 0], 3, pe), ([0, 1], 1, pe)]
    possible_move_dirs[3] = [([1, 0], 3, 1-2 * pe), ([1, 0], 4, pe), ([1, 0], 2, pe)]
    possible_move_dirs[4] = [([1, 0], 4, 1-2 * pe), ([0, -1], 5, pe), ([1, 0], 3, pe)]
    possible_move_dirs[5] = [([0, -1], 5, 1-2 * pe), ([0, -1], 6, pe), ([1, 0], 4, pe)]
    possible_move_dirs[6] = [([0, -1], 6, 1-2 * pe), ([0, -1], 7, pe), ([0, -1], 5, pe)]
    possible_move_dirs[7] = [([0, -1], 7, 1-2 * pe), ([-1, 0], 8, pe), ([0, -1], 6, pe)]
    possible_move_dirs[8] = [([-1, 0], 8, 1-2 * pe), ([-1, 0], 9, pe), ([0, -1], 7, pe)]
    possible_move_dirs[9] = [([-1, 0], 9, 1-2 * pe), ([-1, 0], 10, pe), ([-1, 0], 8, pe)]
    possible_move_dirs[10] = [([-1, 0], 10, 1-2 * pe), ([0, 1], 11, pe), ([-1, 0], 9, pe)]
    possible_move_dirs[11] = [([0, 1], 11, 1-2 * pe), ([0, 1], 0, pe), ([-1, 0], 10, pe)]
    
    P = {}
    if a[0] == FORWARD:
        for possible_move_dir in possible_move_dirs[s[2]]:
            x_hypo = s[0] + possible_move_dir[0][0]
            x = x_hypo if (x_hypo <= X - 1 and x_hypo >= 0) else s[0] # out of the grid cell
            y_hypo = s[1] + possible_move_dir[0][1]
            y = y_hypo if (y_hypo <= Y - 1 and y_hypo >= 0) else s[1] # out of the grid cell
            h = (possible_move_dir[1] + a[1]) % 12
            if possible_move_dir[2] != 0.0:
                P[(x, y, h)] = possible_move_dir[2]

    elif a[0] == BACKWARD:
        for possible_move_dir in possible_move_dirs[s[2]]:
            x_hypo = s[0] - possible_move_dir[0][0]
            x = x_hypo if (x_hypo <= X - 1 and x_hypo >= 0) else s[0] # out of the grid cell
            y_hypo = s[1] - possible_move_dir[0][1]
            y = y_hypo if (y_hypo <= Y - 1 and y_hypo >= 0) else s[1] # out of the grid cell
            h = (possible_move_dir[1] + a[1]) % 12
            if possible_move_dir[2] != 0.0:
                P[(x, y, h)] = possible_move_dir[2]
    else:
        P[s] = 1.0
    return P

def reward(s):
    '''
    given current state, return the reward.
    inputs: s: (x, y, h), current state
    outputs: reward
    '''
    if s[0] in [0, 7] or s[1] in [0, 7]:
        return -100
    elif s[0] == 3 and s[1] in [4, 5, 6]:
        return -10
    elif s[0] == 5 and s[1] == 6:
        return 1
    else:
        return 0

def new_reward(s):
    '''
    This is a new reward function, the target state is (5, 6) with heading = 6.
    '''
    if s[0] in [0, 7] or s[1] in [0, 7]:
        return -100
    elif s[0] == 3 and s[1] in [4, 5, 6]:
        return -10
    elif s[0] == 5 and s[1] == 6 and s[2] == 6:
        return 1
    else:
        return 0

def initialPolicy(S):
    '''
    returns the initial policy pi0.
    inputs: S: all the possible states
    outputs: init_policy: initial policy
    '''
    init_policy = {}
    for s in S:
        goal_dir = [5 - s[0], 6 - s[1]] # the vector that points from current spot to the goal
        if goal_dir == [0, 0]: # already at the goal spot
            init_policy[s] = (STILL, NO_TURN)
        if s[2] in [2, 3, 4]:
            move = FORWARD if goal_dir[0] >= 0 else BACKWARD
        #3theta = np.arctan2(goal_dir[1], goal_dir[0] - move) * 180 / np.pi
        elif s[2] in [5, 6, 7]:
            move = FORWARD if goal_dir[1] <= 0 else BACKWARD
            #theta = np.arctan2(goal_dir[0], goal_dir[1] + move) * 180 / np.pi
        elif s[2] in [8, 9, 10]:
            move = FORWARD if goal_dir[0] <= 0 else BACKWARD
            #theta = np.arctan2(goal_dir[1], goal_dir[0] + move) * 180 / np.pi
        elif s[2] in [11, 0, 1]:
            move = FORWARD if goal_dir[1] >= 0 else BACKWARD
            #theta = np.arctan2(goal_dir[0], goal_dir[1] - move) * 180 / np.pi
        theta = np.arctan2(goal_dir[1], goal_dir[0]) * 180 / np.pi
        angle_diff = s[2] * 30 - (90 - theta)
        if angle_diff > 0 and angle_diff < 180:
            turn = LEFT_TURN
        elif angle_diff == 0 or angle_diff == 180:
            turn = NO_TURN
        else:
            turn = RIGHT_TURN
        init_policy[s] = (move, turn)
    return init_policy

def plotTrajectory(policy, s0, pe=0.0):
    '''
    generates the trajectory given policy, initial state, and error prob.
    inputs: policy: the policy
            s0: initial state
            pe: error prob
    outputs: trajectory list
    '''
    if not (pe >= 0.0 and pe <= 0.5):
        raise ValueError('The error probability must be between 0 and 0.5')
    trajectory = []
    s_current = s0
    while True:
        trajectory.append([s_current, policy[s_current]])
        if policy[s_current] == (0, 0): # no movement, will not move forever
            break
        if s_current[0] == 5 and s_current[1] == 6: # have reached the goal
            break
        possible_next_state = nextState(s_current, policy[s_current], pe)
        s_next = max(possible_next_state, key=possible_next_state.get)
        s_current = s_next
    
    fig = plt.figure(figsize = (X, Y))
    ax = fig.add_subplot(1,1,1)
    plt.xlim((0, X))
    plt.ylim((0, Y))
    plt.grid(True, color = 'k')
    
    edge1 = plt.Rectangle((0,0), 1, 8, color = 'r')
    edge2 = plt.Rectangle((1,0), 7, 1, color = 'r')
    edge3 = plt.Rectangle((7,1), 1, 7, color = 'r')
    edge4 = plt.Rectangle((1,7), 6, 1, color = 'r')
    ax.add_patch(edge1)
    ax.add_patch(edge2)
    ax.add_patch(edge3)
    ax.add_patch(edge4)
    
    yellow = plt.Rectangle((3,4), 1, 3, color = 'gold')
    ax.add_patch(yellow)
    
    goal = plt.Rectangle((5,6), 1, 1, color = 'limegreen')
    ax.add_patch(goal)
    
    plt.plot(s0[0]+0.5, s0[1]+0.5, 'c*', markersize = '40')
    ax.arrow(s0[0]+0.5, s0[1]+0.5, 0.4*np.sin(30*s0[2]*np.pi/180),0.4*np.cos(30*s0[2]*np.pi/180), \
         head_width = 0.1, head_length = 0.2, fc = 'k', ec = 'k')
    
    for i in range(1, len(trajectory)):
        x1 = trajectory[i-1][0][0]
        y1 = trajectory[i-1][0][1]
        x2 = trajectory[i][0][0]
        y2 = trajectory[i][0][1]
        h = trajectory[i][0][2]
        plt.plot([x1+0.5, x2+0.5], [y1+0.5, y2+0.5], '--', color='maroon')
        plt.plot(x2+0.5, y2+0.5, 'bo', markersize = '10')
        ax.arrow(x2+0.5, y2+0.5, 0.4*np.sin(30*h*np.pi/180),0.4*np.cos(30*h*np.pi/180), \
                 head_width = 0.1, head_length = 0.2, fc = 'k', ec = 'k')
        
    plt.show()
    return trajectory

    
def policyEval(S, policy, reward, pe=0.0, gamma=1.0, threshold=0.00001):
    '''
    return the policy evaluation.
    inputs: S: all possible states
            policy: the policy to evaluate, a dictionay S -> A
            reward: the reward function(it is a function)
            pe: error prob.
            gamma: discount factor
            threshold: stop evaluation once value function change is less than threshold for all states
    outputs: dictionay S -> V(value function)
    '''
    if not (pe >= 0.0 and pe <= 0.5):
        raise ValueError('The error probability must be between 0 and 0.5')
    V = {}
    for s in S: # initialize value of all states zero
        V[s] = 0.0
    while True:
        diff = 0
        for s in S:
            v = 0.0
            possible_next_state = nextState(s, policy[s], pe)
            for s_next in possible_next_state.keys():
                p = possible_next_state[s_next]
                v = v + p * (reward(s_next) + gamma * V[s_next])
            diff = max(diff, np.abs(v - V[s])) # how much does value for this state change
            V[s] = v
        if diff < threshold:
            break
    return V

def policyImprovement(S, A, V, reward, pe=0.0, gamma=1.0):
    '''
    return an improved policy.
    inputs: S: all possible states
            A: all possible actions
            V: value function S -> V
            reward: the reward function(it is a function)
            pe: error prob.
            gamma: discount factor
    outputs: policy: the improved policy
        
    '''
    if not (pe >= 0.0 and pe <= 0.5):
        raise ValueError('The error probability must be between 0 and 0.5')
    policy = {}
    for s in S:
        action_values = np.zeros(len(A))
        for i, a in enumerate(A):
            # For each action, look at the possible next states
            possible_next_state = nextState(s, a, pe)
            for s_next in possible_next_state.keys():
                p = possible_next_state[s_next]
                # Calculate the expected value
                action_values[i] += p * (reward(s_next) + gamma * V[s_next])                
        # Find the best action by one-step lookahead
        best_action_index = np.argmax(action_values)
        policy[s] = A[best_action_index]        
    return policy

def policyIteration(S, A, policy, reward, pe, gamma=1.0):
    '''
    performs policy iteration algorithm
    inputs: S:all possible states
            A: all possible actions
            policy: initial policy
            reward: reward function
            pe: error prob.
            gamma: discount factor
    outputs: policy: the policy after iteration
           
    '''
    if not (pe >= 0.0 and pe <= 0.5):
        raise ValueError('The error probability must be between 0 and 0.5')
    while True:
        # Evaluate the current policy
        V = policyEval(S, policy, reward, pe, gamma)       
        # Will be set to false if we make any changes to the policy
        policy_stable = True       
        policy_updated = policyImprovement(S, A, V, reward, pe, gamma)        
        if policy_updated != policy:
            policy_stable = False            
        policy = policy_updated        
        # If the policy is stable we've found an optimal policy. Return it
        if policy_stable:
            return policy, V

if __name__ == '__main__':
    S, A = initialize(X, Y)
    print('There are %d states.' %len(S))
    print('There are %d actions' %len(A))
    
    '''test for transition func.'''
    
    s = (0,3,4)
    s_prime = (0,3,4) 
    for a in A: 
        pr = transitionProb(s, a, s_prime, pe = 0.2) 
        print("action:",a) 
        print("pr:",pr)
    '''test for next state'''
    s = (2,2,2)
    for a in A: 
        possible_next_state = nextState(s, a, 0.2) 
        print("action:",a) 
        print(possible_next_state)
    '''test for initial policy'''
    init_policy = initialPolicy(S)
    print ('The initial policy is: ')
    print(init_policy)
    '''test for trajectory'''
    s0 = (1, 6, 6)
    traj_sample = plotTrajectory(init_policy, s0, pe=0.0)
    print("Trajectory ([state, action]) from %s to the goal is: " % str(s0), traj_sample)
    '''test for 3(e)'''
    s0 = (1, 6, 6)
    trajectory = plotTrajectory(init_policy, s0, 0.0)
    V = policyEval(S, init_policy, reward, gamma=0.9)
    
    for t in trajectory:
        print("The value is:", t[0], V[t[0]])
    
    
    '''test for 3(h)'''
    start = time.time()
    policy_optimal, V_optimal = policyIteration(S, A, init_policy, reward, pe=0.0, gamma=0.9)
    s0 = (1, 6, 6)
    trajectory = plotTrajectory(policy_optimal, s0, pe=0.0)
    V = policyEval(S, policy_optimal, reward, gamma=0.9)
    for t in trajectory:
        print("The value is:", t[0], V[t[0]])
    end = time.time()
    #print("Trajectory ([state, action]) from %s to the goal is: " % str(s0), traj)
    print("Execution time of generating results = %s s" %str(end - start))
    '''test for 5(a)'''
    policy_optimal, V_optimal = policyIteration(S, A, init_policy, reward, pe=0.25, gamma=0.9)
    s0 = (1, 6, 6)
    traj = plotTrajectory(policy_optimal, s0, pe=0.25)
    print("Trajectory ([state, action]) from %s to the goal is: " % str(s0), traj)
    print("Value in state %s is" % str(s0),V_optimal[s0])
    '''test for 5(b), with pe = 0'''
    policy_optimal, V_optimal = policyIteration(S, A, init_policy, new_reward, pe=0.0, gamma=0.9)
    s0 = (1, 6, 6)
    traj = plotTrajectory(policy_optimal, s0, pe=0.0)
    print("Trajectory ([state, action]) from %s to the goal is: " % str(s0),traj)
    print("Value in state %s is" % str(s0),V_optimal[s0])
    '''test for 5(b), with pe = 0.25'''
    policy_optimal, V_optimal = policyIteration(S, A, init_policy, new_reward, pe=0.25, gamma=0.9)
    s0 = (1, 6, 6)
    traj = plotTrajectory(policy_optimal, s0, pe=0.25)
    print("Trajectory ([state, action]) from %s to the goal is: " % str(s0),traj)
    print("Value in state %s is" % str(s0),V_optimal[s0])
    
    policy_optimal, V_optimal = policyIteration(S, A, init_policy, new_reward, pe=0.25, gamma=0.9)
    s0 = (1, 6, 6)
    trajectory = plotTrajectory(policy_optimal, s0, pe=0.25)
    V = policyEval(S, policy_optimal, new_reward, gamma=0.9)
    for t in trajectory:
        print("The value is:", t[0], V[t[0]])
    
```
